```python
import pandas as pd 
import pyreadstat
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
import warnings
```

```python
path_base_censo_2010 = 'df_result_final.sav'
```

```python
df, metadata = pyreadstat.read_sav(path_base_censo_2010, apply_value_formats=False)
```

```python
df
```

```python
df.dtypes
```

```python
df_correlacion = df.corr(method ='pearson')
```

```python
#Visualizacion de correlacion entre variables con matplotlib y seaborn
f, ax = plt.subplots(figsize=(12, 12))
sns.heatmap(df_correlacion , mask=np.zeros_like(df_correlacion , dtype=np.bool), cmap=sns.diverging_palette(220, 10, as_cmap=True),
            square=True, ax=ax)
warnings.filterwarnings("ignore")
plt.savefig("Grafico_de_correlacion.png")
```
