# Union de base de datos Censo y la tipologia de hogar

```python
#Librerias

import pandas as pd 
import pyreadstat
from tqdm import tqdm
```

```python
# Ubicacion del la base de datos censo 2010

path_censo = 'base_censo_2010/CPV2010m_Hogar.sav'

# Ubicacion del la base de datos de tipologia de hogar

path_tipology = 'Tipologia_final.csv'
```

```python
# Lectura de la base de datos Censo 2010 en formato ".sav"

df_censo_2010, metadata_censo_2010 = pyreadstat.read_sav(path_censo, apply_value_formats=False)
```

```python
# Lectura de la base de datos tipologia de hogar en formato ".csv"

df_tipology = pd.read_csv(path_tipology)
```

# Merge entre base de datos

```python
# Crear el id que permitira la union entre base de datos

df_tipology['id_hog']=df_tipology['id'] 
```

```python
# Conocer el tipo de datos  de los id que permitiran la union de las bases de datos
print(df_censo_2010['id_hog'].dtypes)
print(df_tipology['id_hog'].dtypes)
```

```python
# Convertir a un mismo tipo de datos 'int64'

df_censo_2010['id_hog'] = df_censo_2010['id_hog'].astype('int64')
df_tipology['id_hog'] = df_tipology['id_hog'].astype('int64')
```

```python
# Verificar el tipo de datos  de los id que permitiran la union de las bases de datos
print(df_censo_2010['id_hog'].dtypes)
print(df_tipology['id_hog'].dtypes)
```

```python
%%time
df_result_final = pd.merge(df_censo_2010, df_tipology, on= 'id_hog', how='outer' )
```
