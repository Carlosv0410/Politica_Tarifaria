# Union de base de datos Censo y la tipologia de hogar

```python
#Librerias

import pandas as pd 
import pyreadstat
from tqdm import tqdm
```

```python
# Ubicacion del la base de datos censo 2010

path_enemdu = 'data_enemdu/df_resultado.sav'

# Ubicacion del la base de datos de tipologia de hogar

path_tipology = 'Tipologia_final_enemdu.csv'
```

```python
# Lectura de la base de datos Censo 2010 en formato ".sav"

df_enemdu, metadata_censo_enemdu = pyreadstat.read_sav(path_enemdu, apply_value_formats=False)
```

```python
# Lectura de la base de datos tipologia de hogar en formato ".csv"

df_tipology = pd.read_csv(path_tipology)
```

```python
df_enemdu.head()
```

```python
df_tipology.head()
```

# Merge entre base de datos

```python
# Crear el id que permitira la union entre base de datos

df_tipology['id_hog']=df_tipology['id'] 
```

```python
df_enemdu['area'] = df_enemdu['area'].astype(str)
df_enemdu['ciudad'] = df_enemdu['ciudad'].astype(str)
df_enemdu['conglomerado'] = df_enemdu['conglomerado'].astype(str)
df_enemdu['zona'] = df_enemdu['zona'].astype(str)
df_enemdu['sector'] = df_enemdu['sector'].astype(str)
df_enemdu['panelm'] = df_enemdu['panelm'].astype(str)
df_enemdu['vivienda'] = df_enemdu['vivienda'].astype(str)
df_enemdu['hogar'] = df_enemdu['hogar'].astype(str)
```

```python
df_enemdu['id_hog'] = df_enemdu['area']+df_enemdu['ciudad']+df_enemdu['conglomerado']+df_enemdu['zona']+df_enemdu['sector']+df_enemdu['panelm']+df_enemdu['vivienda']+df_enemdu['hogar']
```

```python
# Conocer el tipo de datos  de los id que permitiran la union de las bases de datos
print(df_enemdu['id_hog'].dtypes)
print(df_tipology['id_hog'].dtypes)
```

```python
%%time
df_result_final = pd.merge(df_enemdu, df_tipology, on= 'id_hog', how='outer' )
```
