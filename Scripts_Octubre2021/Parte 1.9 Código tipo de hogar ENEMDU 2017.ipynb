```python
import pandas as pd 
import numpy as np
from tqdm import tqdm
import pyreadstat
```

# Cargar la data con los codigos

```python
df = pd.read_csv("tipology_enemdu.csv")
```

```python
df.head()
```

```python
df = df.rename(columns={'p02':'P01',
                        'p04':'P02',
                        '1.0':'Numero de hijos',
                        '2.0':'Numero de hijas'})
```

# Seleccionar las columnas a utilizarse

```python
# Seleccionar muestra

df = df[['P01', 'P02', 'id', 'Numero de hijos', 'Numero de hijas', 'Parientes', 'No Parientes',
         'Codigo']]
```

```python
df.head()
```

# Tratamiento de datos nulos

```python
# Los nan no son iguales que los ceros

Change = ['Numero de hijos', 'Numero de hijas', 'Parientes', 'No Parientes']
for variable in Change:
    df[variable] = df[variable].fillna(0)
```

# Creación de variable Nh para el conteo total de hijos

```python
# Creamos una columna usefull

df['Nh'] = df['Numero de hijos'] + df['Numero de hijas']
```

```python
df.head()
```

```python
# Funcion

def check(a, b):
    for elemetlist in a:
        if elemetlist in b:
            return True
        return False
```

```python
# Agregamos una columna para reemplazar valor

T = []

for i in tqdm(range(df.shape[0])):
    if check(['2'], list(str(df['Codigo'][i]))):
        if df['Nh'][i] != 0:
            if df['Parientes'][i] != 0:
                if df['No Parientes'][i] != 0:
                    T.append('16')
                else:
                    T.append('6')
            elif df['No Parientes'][i] != 0:
                T.append('10')
            else:
                T.append('2')
        elif df['Parientes'][i] != 0:
            if df['No Parientes'][i] != 0:
                T.append('16')
            else:
                T.append('7')
        elif df['No Parientes'][i] != 0:
            T.append('11')
        else:

            T.append('3')
    else:
        if df['Nh'][i] != 0:
            if df['Parientes'][i] != 0:
                if df['No Parientes'][i] != 0:
                    T.append('16')
                else:
                    if df['P01'][i] == 1.0:
                        T.append('8')
                    elif df['P01'][i] == 2.0:
                        T.append('9')
                    else:
                        T.append('nan')
            elif df['No Parientes'][i] != 0:
                if df['P01'][i] == 1.0:
                    T.append('12')
                elif df['P01'][i] == 2.0:
                    T.append('13')
                else:
                    T.append('nan')
            else:
                if df['P01'][i] == 1.0:
                    T.append('4')
                elif df['P01'][i] == 2.0:
                    T.append('5')
                else:
                    T.append('nan')
        elif df['Parientes'][i] != 0:
            if df['No Parientes'][i] != 0:
                T.append('16')
            else:
                T.append('14')
        elif df['No Parientes'][i] != 0:
            T.append('15')
        else:
            T.append('1')


# Agregar columna en la base de datos

df['Tipology'] = T
```

```python
df
```

```python
dic_tipologia = pd.read_excel('Diccionario tipologia de hogar.xlsx')
```

```python
dic_tipologia
```

```python
df['Tipology'] = df['Tipology'].astype('int64')
dic_tipologia['Tipology'] = dic_tipologia['Tipology'].astype('int64')
```

```python
df = pd.merge(df, dic_tipologia, on= 'Tipology', how='outer' )
```
