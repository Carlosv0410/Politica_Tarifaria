```python
import pandas as pd 
import numpy as np
from tqdm import tqdm
```

```python
df = pd.read_csv('df_result_final.csv')
```

```python
df
```

```python
lista_codigo = list(df['Codigo_Hogar'])
lista_codigo
```

```python
%%time
limpieza_lista = []
for x in lista_codigo:
    y = x.replace('.0', "")
    limpieza_lista.append(y) 
limpieza_lista
```

```python
df['Codigo']=limpieza_lista
```

```python
df
```

```python
%%time
list_code = []
for i  in df.index:
    code = list(df['Codigo'][i])
    list_code.append(code)
list_code
```

```python
df['lista_codigo']= list_code
```

```python
list(df.index)
```

```python
%%time
List_tipology = []
for raw in list(df.index):
    
    #Jefatura masculina
    
    if df['P01'][raw]=='1.0':
        code = df['lista_codigo'][raw]
    
        if '1' in code:
            List_tipology.append('Unipersonal')
```

```python
List_tipology
```

```python
List_tipology = []
for raw in list(df.index) :
    
    #Jefatura masculina
    
    if df['P01'][raw]==1.0:
        code = df['lista_codigo'][raw]
    
        if '1' in code:
            List_tipology.append('Unipersonal')
            
        elif '1' or '2' or '3' in code:
            List_tipology.append('Nuclear Biparental con hijos')
            
            
        elif '1' and '2' in code:
            List_tipology.append('Nuclear Monoparental con jefatura masculina')
            
        else:
            List_tipology.append('no definida todavia')
            
    
    #Jefatura femenina
    elif df['P01'][raw]==2.0:
        code = df['lista_codigo'][raw]
        
        if '1' in code:
            List_tipology.append('Unipersonal')
            
        elif '1' or '2' or '3' in code:
            List_tipology.append('Nuclear Biparental con hijos')
            
            
        elif '1' and '2' in code:
            List_tipology.append('Nuclear Monoparental con jefatura masculina')
            
        else:
            List_tipology.append('no definida todavia')

            
List_tipology    
```
