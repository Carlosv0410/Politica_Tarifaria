```python
# Funcion
def check(a, b):
    for elemetlist in a:
        if elemetlist in b:
            return True
        return False
```

```python
# Agregamos una columna para reemplazar valor

T = []

for i in tqdm(range(df.shape[0])):
    if check(['2'], list(str(df['Codigo'][i]))):
        if df['Nh'][i] != 0:
            if df['Parientes'][i] != 0:
                if df['No Parientes'][i] != 0:
                    T.append('16')
                else:
                    T.append('6')
            elif df['No Parientes'][i] != 0:
                T.append('10')
            else:
                T.append('2')
        elif df['Parientes'][i] != 0:
            if df['No Parientes'][i] != 0:
                T.append('16')
            else:
                T.append('7')
        elif df['No Parientes'][i] != 0:
            T.append('11')
        else:

            T.append('3')
    else:
        if df['Nh'][i] != 0:
            if df['Parientes'][i] != 0:
                if df['No Parientes'][i] != 0:
                    T.append('16')
                else:
                    if df['P01'][i] == 1.0:
                        T.append('8')
                    elif df['P01'][i] == 2.0:
                        T.append('9')
                    else:
                        T.append('nan')
            elif df['No Parientes'][i] != 0:
                if df['P01'][i] == 1.0:
                    T.append('12')
                elif df['P01'][i] == 2.0:
                    T.append('13')
                else:
                    T.append('nan')
            else:
                if df['P01'][i] == 1.0:
                    T.append('4')
                elif df['P01'][i] == 2.0:
                    T.append('5')
                else:
                    T.append('nan')
        elif df['Parientes'][i] != 0:
            if df['No Parientes'][i] != 0:
                T.append('16')
            else:
                T.append('14')
        elif df['No Parientes'][i] != 0:
            T.append('15')
        else:
            T.append('1')


# Agregar columna en la base de datos

df['Tipology'] = T
```
