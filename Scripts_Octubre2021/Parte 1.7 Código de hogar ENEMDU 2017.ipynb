```python
%%time
# Agrupamiento por identificador de hogar y sexo 
data_hijos_enemdu = df_enemdu.groupby(["id_hogar",'p02','p04'])
# tamaño de cada tipo de categoria en variable P01
df_conteo_hijos_enemdu = data_hijos_enemdu.size()
df_hijos_enemdu = df_conteo_hijos_enemdu.reset_index(name='Cuenta genero')
df_genero_hijos_enemdu= df_hijos_enemdu[(df_hijos_enemdu['p04']=='3.0')]
df_conteo_hijos_resultante_enemdu=pd.pivot_table(df_genero_hijos_enemdu,
                                                index='id_hogar',
                                                columns='p02',
                                                values='Cuenta genero')
df_conteo_hijos_resultante_enemdu['id']=df_conteo_hijos_resultante_enemdu.index
df_conteo_hijos_resultante_enemdu
```

```python
%%time
data_parientes_enemdu = df_enemdu.groupby(["id_hogar",'p04'])
conteo_parientes_enemdu = data_parientes_enemdu.size()
df_parientes_enemdu =conteo_parientes_enemdu.reset_index(name='Cuenta')
df_parientes_enemdu = df_parientes_enemdu[(df_parientes_enemdu['p04']=='4.0')|(df_parientes_enemdu['p04']=='5.0')|(df_parientes_enemdu['p04']=='6.0')|(df_parientes_enemdu['p04']=='7.0')]
df_parientes_enemdu['Parientes_resultantes'] ='Parientes'
df_parientes_resultante_enemdu = pd.pivot_table(df_parientes_enemdu,
                                               index = 'id_hogar',
                                               columns = "Parientes_resultantes" ,
                                               values = "Cuenta",
                                               aggfunc = "sum")
df_parientes_resultante_enemdu['id']=df_parientes_resultante_enemdu.index
df_parientes_resultante_enemdu
```

```python
%%time
data_no_parientes_enemdu = df_enemdu.groupby(["id_hogar",'p04'])
conteo_no_parientes_enemdu = data_no_parientes_enemdu.size()
df_no_parientes_enemdu = conteo_no_parientes_enemdu.reset_index(name='Cuenta')
df_no_parientes_enemdu = df_no_parientes_enemdu[(df_no_parientes_enemdu['p04']=='8.0')|(df_no_parientes_enemdu['p04']=='9.0')|(df_no_parientes_enemdu['p04']=='10.0')|(df_no_parientes_enemdu['p04']=='11.0')]
df_no_parientes_enemdu['No_Parientes_resultantes'] ='No Parientes'
df_no_parientes_resultante_enemdu = pd.pivot_table(df_no_parientes_enemdu,
                                                    index = 'id_hogar',
                                                    columns = "No_Parientes_resultantes",
                                                    values = "Cuenta",
                                                    aggfunc = "sum")
df_no_parientes_resultante_enemdu['id']=df_no_parientes_resultante_enemdu.index
df_no_parientes_resultante_enemdu
```
