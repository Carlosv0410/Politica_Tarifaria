# Limpiar las codificaciones y transformar a listas sus valores

## Librerias

```python
import pandas as pd 
import numpy as np
from tqdm import tqdm
```

```python
df = pd.read_csv('df_result_final_enemdu.csv')
```

```python
df
```

```python
lista_codigo = list(df['Codigo Hogar'])
lista_codigo
```

```python
limpieza_lista = []
for x in lista_codigo:
    y = x.replace('.0', "")
    limpieza_lista.append(y) 
limpieza_lista
```

```python
df['Codigo']=limpieza_lista
```

```python
df
```

```python
list_code = []
for i  in df.index:
    code = list(df['Codigo'][i])
    list_code.append(code)
list_code
```

```python
df['lista_codigo']= list_code
```

```python
df
```

```python
df['Tipology'] = df['lista_codigo']
```
